```python
import numpy as np
```

```python
A = np.random.randint(1,100,9)
A = A.reshape(3,3)
print("Martix A:\n",A)
```

```python
col = np.array([[4],[2],[5]])
row = np.array([1,2,3])
print("Vector column:\n",col)
print("Vector row:\n",row)
print("\nMatrix A\n",A)
print("\nMatrix A + 10:\n", A+10)
print("\nMatrix A + row vector\n", A+row)
print("\nMatrix A + column vector\n", A+col)
```
