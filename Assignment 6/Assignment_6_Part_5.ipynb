```python
import numpy as np
```

***PART 1***



```python
A = np.random.randint(1,25,5)
B = np.random.randint(26,100,5)
```

```python
print("MATRIX A:\n",A)
print("MATRIX B:\n",B)
```

```python
print("Addition of Matrix A and B:\n",A+B)
print("\nSubtraction of Matrix A and B:\n",A-B)
print("\nMultiplication of Matrix A and B:\n",A*B)
print("\nDivision of Matrix A and B:\n",A/B)
```

***PART*** ***2***

```python
X = np.random.randint(1,100,9)
X = X.reshape(3,3)
```

```python
Square_Root = np.sqrt(X)
print("Square Root of Matrix X:\n",Square_Root)
Exp = np.exp(X)
print("\nExponential of Matrix X:\n",Exp)
Sin = np.sin(X)
print("\nSin of Matrix X:\n",Sin)
```
