```python
class BankAccount ():
  def __init__ (self, acc_no, bal):
    self.__acc_no = acc_no
    self.__bal = int(bal)


  def deposite (self):
    acc_no = input("Enter Your Account Number:\n\n")
    if acc_no == self.__acc_no:
      print(f'\n--Welcome to your Account--\n\nYour present Balance: {self.__bal}\n')
      amount = int(input("Enter Amount to deposite: "))
      self.__bal += amount
      print(f'\nYour updated balance: {self.__bal}')

    else :
      print("\nInvalid Account Number!!!")


  def withdraw (self):
    acc_no = input("Enter Your Account Number:\n\n")
    if acc_no == self.__acc_no: # Fixed attribute name
      print(f'\n--Welcome to your Account--\n\nYour present Balance: {self.__bal}\n')
      amount = int(input("Enter Amount to withdraw: "))
      if self.__bal > 0 and self.__bal >= amount:
        self.__bal -= amount
        print(f'\nYour updated balance: {self.__bal}')
      else:
        print("\nInsufficient Balance!!!")
    else:
      print("\nInvalid Account Number!!!")


  def get_balance (self):
    acc_no = input("Enter Your Account Number:\n\n")
    if acc_no == self.__acc_no: # Fixed attribute name
      print(f'\n--Welcome to your Account--\n\nYour present Balance: {self.__bal}\n')
    else:
      print("\nInvalid Account Number!!!")
```

```python
my_acc = BankAccount ("A098", 9990)
```

```python
my_acc.get_balance()
```

```python
my_acc.deposite()
```

```python
my_acc.withdraw()
```
