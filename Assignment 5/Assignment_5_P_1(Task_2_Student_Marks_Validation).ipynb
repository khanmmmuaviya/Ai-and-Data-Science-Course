```python
class Student ():
  def __init__(self, name, roll_no, marks):
    self.__name = name
    self.__roll_no = roll_no
    self.__marks = int(marks)

  def setter (self):
    new_marks = int(input("Enter new marks: "))
    if new_marks >= 0 and new_marks <= 100:
      self.__marks = new_marks
      print(f"Updated Marks: {new_marks}")
    else:
      print("Marks limit exceeded!!!")


  def getter (self):
    print(f"Name: {self.__name}")
    print(f"Roll No: {self.__roll_no}")
    print(f"Marks: {self.__marks}")
```

```python
st_1 = Student('Muaviya', 11, 88)
```

```python
st_2 = Student('Umer', 12, 80)
```

```python
st_1.getter()
```

```python
st_1.setter()
```

```python
st_2.getter()
```

```python
st_2.setter()
```
