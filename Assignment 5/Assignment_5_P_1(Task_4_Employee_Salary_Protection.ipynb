```python
class Employee ():
  def __init__ (self, name, salary):
    self.__name = name
    self.__salary = int(salary)

  def getter (self):
    return self.__name

  def setter (self):
    if self.getter ():
      print(f"The current salary of {self.getter()} is {self.__salary}")
      new_sal = int(input(f"\nEnter new salary for {self.getter()}: "))
      self.__salary = new_sal
      print (f"\nThe new ssalary of {self.getter()} is {self.__salary}")

  def show_details (self):
    print(f"Name: {self.__name}")
    print(f"\nSalary: {self.__salary}")


```

```python
emp = Employee("John", 50000)
```
