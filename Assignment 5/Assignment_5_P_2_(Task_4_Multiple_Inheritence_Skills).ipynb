```python
class Father ():
  def __init__(self, age, skill):
    self.f_skill = skill
    self.f_age = age


  def get_skill(self):
    print(f"Father's skill is {self.skill}")


class Mother ():
  def __init__(self, age, skill):
    self.m_skill = skill
    self.m_age = age


  def get_skill(self):
    print(f"Mother's skill is {self.skill}")


class Child (Father, Mother):
  def __init__(self, f_age, m_age, f_skill, m_skill):
    Father.__init__(self, f_age, f_skill)
    Mother.__init__(self, m_age, m_skill)

  def get_skill(self):
    print(f"The child's skills are {self.f_skill} and {self.m_skill}")
```

```python
ch = Child(49, 38, "Scientist", "Freelancer")
```
