```python
class GrandParents ():
  def __init__(self, family_name):
    self.family_name = family_name

  def get_family_name(self):
    return (f"The family name is {self.family_name}")

class Parents (GrandParents):
  def __init__(self, family_name, occupation):
    super().__init__(family_name)
    self.occupation = occupation

  def get_occupation(self):
    return f"Their occupation is {self.occupation}"

class Child (Parents):
  def __init__(self, family_name, occupation, hobby):
    super().__init__(family_name, occupation)
    self.hobby = hobby

  def get_hobby(self):
    return f"The child's hobby is {self.hobby}."

  def about_child (self):
    print (f"{self.get_family_name()}\n{self.get_occupation()}\n{self.get_hobby()}")
```
