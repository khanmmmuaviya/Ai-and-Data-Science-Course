```python
class Vehical :
  def __init__(self, brand, name, model):
    self.brand = brand
    self.name = name
    self.model = model

  def info(self):
    print(f"Brand: {self.brand}\nCar Name: {self.name}\nModel: {self.model}")


class Car (Vehical):
  def __init__(self, brand, name, model, seats):
    super().__init__(brand, name, model)
    self.seats = seats

  def info(self):
    super().info()
    print(f"Seats: {self.seats}")
```

```python
car = Car("Mazda", "RX7", 2012, 2)
```
