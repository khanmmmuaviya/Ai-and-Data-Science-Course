```python
class ShoppingCart ():
    def __init__(self, items=[]):
        self.__items = items


    def add_item(self, item):
        self.__items.append(item)

    def remove_item(self, item):
        self.__items.remove(item)

    def get_items(self):
        return self.__items

```

```python
cart = ShoppingCart(["apples", "biscuits", "chips", "jellys", "groceries", "toys", "Smoking essentials", "BBQ essentials"] )
```

```python
cart.get_items()
```

```python
cart.add_item("frozen nuggets")
```

```python
cart.get_items()
```

```python
cart.remove_item("Smoking essentials")
```
